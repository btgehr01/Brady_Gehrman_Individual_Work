```python

# Build and train a KNN Regression model. 
# Vary the parameter K and analyze the results by identifying cases of overfitting and underfitting. 
# Select the optimal value of K and justify your choice.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

training_data = pd.read_csv('datasets_Hwk1/wine_train.csv')
test_data = pd.read_csv('datasets_Hwk1/wine_test.csv')

x_train = training_data.iloc[:, :-1].values

y_train = training_data["quality"].values

x_test = test_data.iloc[:, :-1].values

y_test = test_data["quality"].values

num_of_neighbors = np.arange(1, 81)

training_set_scores = []
test_set_scores = []

for k in num_of_neighbors:
    reg = KNeighborsRegressor(n_neighbors=k)
    reg.fit(x_train, y_train)
    training_set_scores.append(reg.score(x_train, y_train))
    test_set_scores.append(reg.score(x_test, y_test))

print(training_set_scores)
print(test_set_scores)
plt.figure(figsize=(10, 6))
plt.plot(num_of_neighbors, training_set_scores, label='Training Set')
plt.plot(num_of_neighbors, test_set_scores, label='Testing Set')
plt.xlabel('K (# Neighbors)')
plt.ylabel('Score')
plt.title('KNN Regression: Training vs. Testing Set Performance')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# Build and train a Ridge Regression model 
# Vary the constraint parameter alpha (learning rate) and analyze the results by identifying cases of overfitting and underfitting.
# Select the optimal value of alpha

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

training_data = pd.read_csv('datasets_Hwk1/wine_train.csv')
test_data = pd.read_csv('datasets_Hwk1/wine_test.csv')

x_train = training_data.iloc[:, :-1].values

y_train = training_data["quality"].values

x_test = test_data.iloc[:, :-1].values

y_test = test_data["quality"].values

alpha_values = [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]

training_set_scores = []
test_set_scores = []

for alpha in alpha_values:
    ridge = Ridge(alpha=alpha)
    ridge.fit(x_train, y_train)
    training_set_scores.append(ridge.score(x_train, y_train))
    test_set_scores.append(ridge.score(x_test, y_test))

print(test_set_scores)
plt.figure(figsize=(10, 6))
plt.semilogx(alpha_values, training_set_scores, label='Training Set')
plt.semilogx(alpha_values, test_set_scores, label='Testing Set')
plt.xlabel('Alpha (Learning Rate)')
plt.ylabel('Score (R^2)')
plt.title('Ridge Regression: Training vs. Testing Set Performance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Build and train a LASSO Regression model (tries to set parameters to equal 0, feature selection). 
# Vary the constraint parameter alpha and analyze the results by identifying cases of overfitting and underfitting. 
# Select the optimal value of aplha

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

training_data = pd.read_csv('datasets_Hwk1/wine_train.csv')
test_data = pd.read_csv('datasets_Hwk1/wine_test.csv')

x_train = training_data.iloc[:, :-1].values

y_train = training_data["quality"].values

x_test = test_data.iloc[:, :-1].values

y_test = test_data["quality"].values

alpha_values = [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]

training_set_scores = []
test_set_scores = []
num_features_used = []

for alpha in alpha_values:
    lasso = Lasso(alpha=alpha)
    lasso.fit(x_train, y_train)
    training_set_scores.append(lasso.score(x_train, y_train))
    test_set_scores.append(lasso.score(x_test, y_test))
    num_features_used.append(np.sum(lasso.coef_ != 0))
    
print(test_set_scores)
plt.figure(figsize=(10, 12))

plt.subplot(2, 1, 1)
plt.semilogx(alpha_values, training_set_scores, label='Training Set')
plt.semilogx(alpha_values, test_set_scores, label='Testing Set')
plt.xlabel('Alpha (Learning Rate)')
plt.ylabel('Score (R^2)')
plt.title('Lasso Regression: Training vs. Testing Set Performance')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.semilogx(alpha_values, num_features_used, color='purple')
plt.xlabel('Alpha (Learning Rate)')
plt.ylabel('Number of Used features')
plt.title('Lasso Regression: Number of Non-Zero Coefficients vs. Alpha')
plt.grid(True)
plt.tight_layout()
plt.show()
```
