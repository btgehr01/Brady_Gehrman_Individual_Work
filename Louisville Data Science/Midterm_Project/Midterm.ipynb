```python
#Load the .csv file into a pandas dataframe
import pandas as pd
# use the pandas read_csv method
df_prob1 = pd.read_csv("MetObjects_Subset.csv")
display(df_prob1)
```

Question 2:

```python
# copy df_prob1
df_prob2 = df_prob1.copy()
# find which columns are "empty" using boolean indexing
emptyColumns = df_prob2.columns[df_prob2.isna().all()]
print(emptyColumns)
# drop the empty columns
df_prob2 = df_prob2.drop(columns=emptyColumns)
# display the new pandas DF
display(df_prob2)
```

Question 3:

```python
df_prob3 = df_prob2.copy()
# keep the rows within the DF that have an "Is Public Domain" value of True
df_prob3 = df_prob3[df_prob3["Is Public Domain"] == True]
display(df_prob3)
```

Question 4:

```python
df_prob4 = df_prob3.copy()
# fill all NA artist role values with an empty string because some are NA
df_prob4["Artist Role"] = df_prob4["Artist Role"].fillna("")
# keep the rows within the DF that have an artist role that contains "Maker"
df_prob4 = df_prob4[df_prob4["Artist Role"].str.contains("Maker")]
display(df_prob4)
```

Question 5:

```python
# define a function that will be applied to each row of the DF and output a value (DecadesToMake)

def outputDecades(rowObject):
    if pd.isna(rowObject["Object Begin Date"]) or pd.isna(rowObject["Object End Date"]):
        return 0
    else:
        return (rowObject["Object End Date"] - rowObject["Object Begin Date"]) // 10
    
    

df_prob5 = df_prob4.copy()
# apply the outputDecades function to each row in the DF and create a new column with those values named "DecadesToMake"
df_prob5["DecadesToMake"] = df_prob5.apply(outputDecades, axis=1)
display(df_prob5)
```

Question 6:

```python
import matplotlib.pyplot as plt

# strip the classification value from each row
differentClassificationValues = df_prob5["Classification"].values

# define a new DF with two columns ("Word" and "Count")
columns = ["Word", "Count"]
newDF = pd.DataFrame(columns=columns)

# loop through each classification value from the previous DF
for value in differentClassificationValues:
    # increment the "Count" value if the classification value is already within the newDF (present within the "Word" column)
    if value in newDF["Word"].values:
        indexOfWord = newDF.index[newDF["Word"] == value]
        previousCount = newDF.loc[indexOfWord, "Count"].iloc[0]
        newDF.loc[indexOfWord, "Count"] = previousCount + 1
    # add the classification value to the newDF if it isn't present within the "Word" column already
    else:
        newDF.loc[len(newDF)] = {"Word": value, "Count": 1}

# sort the newDF by count and strip the top 10
sortedDF = newDF.sort_values(by="Count", ascending=False)
top10DF = sortedDF.head(10)

# generate the bar plot using matplotlib and show it
plt.figure(figsize=(14, 8)) 
plt.bar(top10DF["Word"].values, top10DF["Count"].values)

plt.xlabel("Classification")
plt.ylabel('Counts')
plt.title("Item Classification Counts")
plt.tight_layout() 

plt.show()
```
